# Feature  Scaling

```python
Med[['sex', 'smoker', 'region']] = Med[['sex', 'smoker', 'region']].astype('category')
Med.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(Med.sex.drop_duplicates())
Med.sex = label.transform(Med.sex)
label.fit(Med.smoker.drop_duplicates())
Med.smoker = label.transform(Med.smoker)
label.fit(Med.region.drop_duplicates())
Med.region = label.transform(Med.region)
Med.dtypes
```

```python
Med.head()
```

```python
fig = plt.gcf();
fig.set_size_inches(12, 6); 
sns.heatmap(Med.corr(), annot=True)
```

```python
Med['charges']=np.log10(Med['charges'])#This distribution is right-skewed previously. To make it closer to normal we can apply natural log
```

```python
fig = plt.gcf();
fig.set_size_inches(6, 6);
sns.distplot(Med['charges']);
```

```python
import seaborn as sns
sns.boxplot(data=Med['charges']);
```

# Simple-Linear Regression

```python
from sklearn.model_selection import train_test_split as holdout
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x =Med['smoker']
y =Med['charges']
```

```python
x_train, x_test, y_train, y_test = holdout(x, y, test_size=0.2, random_state=100)
Lin_reg = LinearRegression()
Lin_reg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred=Lin_reg.predict(np.array(x_test).reshape(-1,1))
```

```python
print(Lin_reg.intercept_)
print(Lin_reg.coef_)
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, y_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
sns.regplot(y_test,y_pred)
```

# Multiple-Linear Regression

```python
from sklearn.model_selection import train_test_split as holdout
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x2 = Med[['age','sex','bmi','children','smoker','region']]
y2 = Med['charges']
```

```python
x2_train, x2_test, y2_train, y2_test = holdout(x2, y2, test_size=0.2, random_state=101)
Lin_reg = LinearRegression()
Lin_reg.fit(x2_train, y2_train)
y_pred2=Lin_reg.predict(x2_test)
```

```python
print(Lin_reg.intercept_)
print(Lin_reg.coef_)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y2_test,y_pred2))
```

```python
print('MAE:',metrics.mean_absolute_error(y2_test, y_pred2))
print('MSE:',metrics.mean_squared_error(y2_test, y_pred2))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y2_test, y_pred2)))
```

```python
sns.regplot(y2_test,y_pred2)
```

```python
sns.distplot(y2_test-y_pred2)
```

# Polynomial - Regression

```python
from sklearn.preprocessing import PolynomialFeatures
x = Med.drop(['charges', 'sex', 'region'], axis = 1)
y = Med.charges
```

```python
pol = PolynomialFeatures (degree = 2)
x_pol = pol.fit_transform(x)
x3_train, x3_test, y3_train, y3_test = holdout(x_pol, y, test_size=0.2, random_state=101)
Pol_reg = LinearRegression()
Pol_reg.fit(x3_train, y3_train)
y_train_pred = Pol_reg.predict(x3_train)
y_test_pred = Pol_reg.predict(x3_test)
```

```python
print(Pol_reg.intercept_)
print(Pol_reg.coef_)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y3_train,y_train_pred))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y3_test,y_test_pred))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y3_train,y_train_pred,
          c = 'green', marker = 'o', s = 35, alpha = 0.5,
          label = 'Train data')
plt.scatter(y3_test,y_test_pred,
          c = 'orange', marker = 'o', s = 35, alpha = 0.7,
          label = 'Test data')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.legend(loc = 'lower right')
```

```python
##Evaluating the performance of the algorithm(train)
print('Mean Absolute Error:', metrics.mean_absolute_error(y3_train, y_train_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y3_train, y_train_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y3_train, y_train_pred)))
```

```python
##Evaluating the performance of the algorithm (test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y3_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y3_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y3_test, y_test_pred)))
```

```python
sns.regplot(y3_train,y_train_pred)
```

```python
sns.regplot(y3_test,y_test_pred)
```

```python
sns.distplot(y3_test-y_test_pred)
```

# Decision Tree Regression

```python

from sklearn.tree import DecisionTreeRegressor 
X = Med.drop(['charges', 'sex', 'region'], axis = 1)
y = Med.charges
x4_train, x4_test, y4_train, y4_test = holdout(X, y, test_size=0.2, random_state=101)
regressor = DecisionTreeRegressor() 

regressor.fit(x4_train,y4_train) 
y_pred4=regressor.predict(x4_test)

from sklearn.metrics import r2_score
print(r2_score(y4_test,y_pred4))
```

```python
sns.regplot(y4_test,y_pred4)
```

```python
##Evaluating the performance of the algorithm (test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y4_test, y_pred4))
print('Mean Squared Error:', metrics.mean_squared_error(y4_test, y_pred4))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y4_test, y_pred4)))
```

Conclusion: like when previously noticed smoking is the greatest factor that affects medical cost charges, then it's bmi and age. Polynomial Regression turned out to be the best model
